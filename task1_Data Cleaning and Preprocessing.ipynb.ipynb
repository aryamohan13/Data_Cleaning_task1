```python
import pandas as pd
data = pd.read_csv("netflix_titles.csv")  
data.head()
```

```python
data.isnull().sum()
```

```python
data['director']=data['director'].fillna('unknown')
data['cast']=data['cast'].fillna('prefered not to say')
data['country']=data['country'].fillna('Concealed')
data['rating']=data['rating'].fillna('Nothing to Say')
data['duration']=data['duration'].fillna('Unpredictable')
data.head(10)
```

Dates Formating:

```python
data['date_added'] = data['date_added'].str.strip()
data['date_added'] = pd.to_datetime(data['date_added'], format='%B %d, %Y', errors='coerce')
data['date'] = data['date_added'].dt.strftime('%d-%m-%Y')
```

Droping of Duplicates:

```python
data.drop(columns=['date_added'],inplace=True)
data.tail(5)
```

Standarization For country:

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['country_encoded'] = le.fit_transform(data['country'])
data.head()
```

```python
data.to_csv("cleaned data.csv",index=False)
```
